# Line Chart

```python
#it is a graph that displays information as a series of data points called 'markers'
#connected by straight line segments
#It is commonly used to visualize a trend in data over intervals of time.
#we will do trend analysis
```

```python
import matplotlib.pyplot as plt
month=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales=[1200,1500,1300,2400,5600,9800,6700,2340,4567,8764,3900,5000]

plt.figure(figsize=(10,6))
plt.plot(month,sales,marker='o',linestyle='-',color='b',label='Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly sales')
plt.legend()
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [1200, 1500, 1300, 2400, 5600, 9800, 6700, 2340, 4567, 8764, 3900, 5000]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(month, sales, marker='o', linestyle='-', color='b', label='Sales')  # Fixed variable and style
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')  # Fixed typo in title
plt.legend()
plt.grid()

# Show plot
plt.show()
```
