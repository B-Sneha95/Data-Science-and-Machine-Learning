# Part C: OOP Implementation 

```python
# a) SocialPost class with encapsulation
class SocialPost:
    def __init__(self, post_id, content, likes, comments):
        self.post_id = post_id
        self.content = content
        self.__likes = likes
        self.__comments = comments

    # Getter for likes
    def get_likes(self):
        return self.__likes

    # Setter for likes
    def set_likes(self, new_likes):
        if new_likes >= 0:
            self.__likes = new_likes

    # Getter for comments
    def get_comments(self):
        return self.__comments

    # Setter for comments
    def set_comments(self, new_comments):
        if new_comments >= 0:
            self.__comments = new_comments

    # Display method for polymorphism
    def display(self):
        print(f"Post ID: {self.post_id}")
        print(f"Content: {self.content}")
        print(f"Likes: {self.__likes}")
        print(f"Comments: {self.__comments}")
        
```

```python
# b) Author subclass inheriting from SocialPost
class Author(SocialPost):
    def __init__(self, post_id, content, likes, comments, author_name, author_rating):
        super().__init__(post_id, content, likes, comments)  # call parent constructor
        self.author_name = author_name
        self.author_rating = author_rating

    # Overriding display() => Polymorphism
    def display(self):
        print(f"Author: {self.author_name}")
        print(f"Author Rating: {self.author_rating}")
        super().display()  # reuse the parent's display


# Test the classes
post1 = SocialPost("POST001", "Excited to launch my blog!", 150, 20)
post2 = Author("POST002", "Check out my tech review!", 300, 45, "Sneha", 4.7)

# Demonstrate polymorphism
print("\n--- SocialPost Display ---")
post1.display()

print("\n--- Author Display (Polymorphism) ---")
post2.display()
```

# Part D: Data Visualization 

```python
import matplotlib.pyplot as plt
import seaborn as sns

# a) Line chart showing daily post frequency
daily_counts = df['Date_Posted'].value_counts().sort_index()
plt.figure(figsize=(10,5))
daily_counts.plot(kind='line')
plt.title("Daily Post Frequency")
plt.xlabel("Date")
plt.ylabel("Number of Posts")
plt.grid(True)
plt.show()
```

```python
# b) Stacked bar chart of average engagement (Likes+Shares+Comments) by category
df['Total_Engagement'] = df['Likes'] + df['Shares'] + df['Comments']
avg_engagement = df.groupby('Category')[['Likes', 'Shares', 'Comments']].mean()
avg_engagement.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Average Engagement by Category")
plt.ylabel("Average Count")
plt.show()
```

```python
# c) Seaborn heatmap of correlation matrix
plt.figure(figsize=(8,6))
sns.heatmap(df[['Likes', 'Shares', 'Comments', 'Post_Length', 'Sentiment_Score']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
