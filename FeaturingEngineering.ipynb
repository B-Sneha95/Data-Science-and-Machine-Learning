# 3.Outlier Detection

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Boxplot for detecting outliers in the "Price" column
plt.figure(figsize=(3, 2))
sns.boxplot(x=df["cost"])
plt.title("Boxplot for Price")
plt.show()
```

```python
# Boxplot for detecting outliers in the "Size" column
plt.figure(figsize=(3,2))
sns.boxplot(x=df["location_area"])
plt.title('Boxplot for Area size')
plt.show()
```

```python
#b.Handle the outliers by using the Interquartile Range (IQR) method.
# Function to handle outliers using the IQR method
def handle_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)  # 25th percentile
    Q3 = df[column].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1  # Interquartile Range
    
    # Bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Remove outliers
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    # Return the cleaned DataFrame
    return df

df = handle_outliers_iqr(df, "cost")           
df = handle_outliers_iqr(df, "location_area")


# Check the cleaned DataFrame
print(df)

```

# 6.Correlation Ananlysis

```python
# Step 4: Correlation Analysis and Heatmap
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure all non-numeric columns are dropped or converted to numeric before calculating correlation
df_numeric = df.select_dtypes(include=["number"])

# Calculate the correlation matrix
correlation_matrix = df_numeric.corr()

# Plot a heatmap to visualize correlations
plt.figure(figsize=(5, 4))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 6a: Drop features with low correlation to the target variable "cost"
# Define a correlation threshold (e.g., 0.2)
# For instance, if encoding was done earlier, define df_encoded
df_encoded = pd.get_dummies(df, columns=['furniture_status'], drop_first=True)

# Calculate the correlation matrix
correlation_matrix = df_encoded.corr()

# Define a correlation threshold
threshold = 0.2
target_variable = "cost"

# Identify low-correlation features
low_correlation_features = correlation_matrix[target_variable][correlation_matrix[target_variable].abs() < threshold].index.tolist()

# Drop low-correlation features
df_reduced = df_encoded.drop(columns=low_correlation_features)
```

```python
# Print results
print(f"Features dropped due to low correlation: {low_correlation_features}")
print("\nDataFrame after dropping low-correlation features:")
print(df_reduced.head())
```

```python
# Explanation for dropping low-correlation features
#"""
Why drop features with low correlation?

1. **Noise Reduction**: Features with low correlation provide little predictive power and add noise, which can reduce model performance.
2. **Simplified Model**: Removing irrelevant features reduces complexity, making models faster to train and easier to interpret.
3. **Avoid Overfitting**: Low-correlation features can lead to overfitting, where the model learns patterns that do not generalize well to unseen data.
4. **Improved Interpretability**: Models with fewer, relevant features are easier to understand and explain.
#"""
```
