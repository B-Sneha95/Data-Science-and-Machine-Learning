# Skewness and symmetry

```python
#Skewness is a statistical measure that describes the asymmetry of a dataset's distribution around its mean. 
#It helps to identify whether the data distribution is symmetrical or tilted to one side.
#to detect the outliers and understand  the spread of skewness of a data 
#Skewness = 0: Perfect symmetry.
#The data is evenly distributed around the mean.
#Skewness > 0: Right-skewed (positive).
#The right tail (larger values) is longer or fatter than the left tail.
#The mean is typically greater than the median.
#Example: Income distribution in many countries.
#Skewness < 0: Left-skewed (negative).
#The left tail (smaller values) is longer or fatter than the right tail.
#The mean is typically less than the median.
#Example: Scores where most students perform well but a few score very low.
#Why is Skewness Important?
#Skewness gives insights into the nature of the data, helping analysts decide on appropriate statistical methods and transformations. 
#For example, data with high skewness might need transformation (e.g., logarithmic) to meet the assumptions of statistical tests or models.
```

```python
#Sample data
import pandas as pd
data={'jobrole':['Engineer','Engineer','Engineer','Engineer','manager','manager'],
      'Salary':[100,30000,40000,50000,89000,760000]}
df=pd.DataFrame(data)
df
```

```python
df.plot(kind='box')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(x='jobrole',y='Salary',data=df)
plt.title('Salary distribution')
plt.xlabel('Jobrole')
plt.ylabel('Salary')
plt.show()
```

# Violen Plot

```python
#A violen plot combines aspects of  boxplot with kernel density plot
#showing the density of the data at different values
#USE:To visualise the distribution of continuos variable across different categories
#when you want to see the probability density of data along with the summary statistics
#eg:Suppose you want to analyse the distribution of test scores
```

```python
#used to visualise the distribution of continuos variable accross different categories
#when you want to probability density of data along with the summary statistics
#eg:suppose if you want to analyse the distribution of test scores across different classes
```

```python
data1={'class':['A','B','A','B','A','A','B'],
       'testscore':[50,90,89,95,76,97,80]}

df=pd.DataFrame(data1)
df
```

```python
sns.violinplot(x='class',y='testscore',data=data1)
plt.title('TestScores')
plt.xlabel('class')
plt.ylabel('Scores')
plt.show()
```

```python
#To visualise the correlation between variables in a dataset
#T display matrix lke data
#eg:suppose if you have a dataset with various numerical features and if you want to visualise the correlation between them
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data=np.random.rand(10,12)
df=pd.DataFrame(data,columns=[f'Var{i}' for i in range (1,13)])
#compute correlation matrix
corr=df.corr()

#Heat Map
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Correlation heat map')
plt.show()
```

# Swarm plot

```python
#A sworm plot is a type of scatter plot that shows one dimensional data points grouped together avoiding overlap
#when you want to see all the individual data points without overlap
```

```python
#eg:suppode if you have test scores of students from different classes and you want to see the distribution of these scores
```

```python
data={'class':['A','B','B','A','A','B','A'],'test_score':[90,89,89,78,97,99,98]}
df=pd.DataFrame (data)
df
```

```python
sns.swarmplot(x='test_score',y='class',data=df)
plt.xlabel('Scores')
plt.ylabel('class')
plt.title('Distibution of test scores')
plt.show()
```

# Density plot

```python
#A density plot is a smoothed ,continuous version of a histogram estimated from the data
#It shows the distribution of a continuous variable
#eg:to visualise the distribution of continuous variable
#when you want to smooth out the noise of a histogram
#suppose you want to analyse the distribution of test scores for student
```

```python
data={'testscore':[78,87,89,98,90,95,88,93,90]}
df=pd.DataFrame(data)
df
```

```python
sns.kdeplot(df['testscore'],shade=True)
plt.title('Density Plot')
plt.xlabel('Test Score')
plt.ylabel('Density')
plt.show()
```
